```python
"""Softmax."""

import numpy as np

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    e = np.exp(np.array(x))
    return e / np.sum(e, axis=0)

scores = np.array([3.0, 1.0, 0.2])
print(softmax(scores))

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

When scores are multipled by 10, before going through the softmax, the probabilites get closer to the extreme values [0, 1].

When scores are divided by 10, the probabilites get closer to the uniform distribution, and therefore similar in value
